```python
# 예상 연봉을 구간별로 나누기
salary_bins = pd.cut(data['EstimatedSalary'], bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000])
salary_bin_counts = salary_bins.value_counts().sort_index()

# 구간별 빈도수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=salary_bin_counts.index, y=salary_bin_counts.values)
plt.xticks(rotation=45)
plt.title('예상 연봉 구간별 분포')
plt.xlabel('Estimated Salary Range')
plt.ylabel('Frequency')
plt.show()
```

```python
# 예상 연봉을 구간별로 나누기
salary_bins = pd.cut(data['EstimatedSalary'], bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000])
salary_bin_counts = salary_bins.value_counts().sort_index()

# 구간별 빈도수 시각화
plt.figure(figsize=(12, 7))
ax = sns.barplot(x=salary_bin_counts.index.astype(str), y=salary_bin_counts.values, palette='viridis')

# 각 막대에 데이터 레이블 추가
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.xticks(rotation=45)
plt.title('예상 연봉 구간별 분포', fontsize=15, fontweight='bold')
plt.xlabel('Estimated Salary Range', fontsize=13)
plt.ylabel('Frequency', fontsize=13)

# 그리드 추가
plt.grid(axis='y')

# 그래프 표시
plt.show()
```
