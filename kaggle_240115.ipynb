## 고객 분포 시각화

### 지역별 고객 분포

```python
# 지역별 고객 분포
plt.figure(figsize=(10, 6))
sns.countplot(x='Geography', data=data, palette='Set2')
plt.title('지역별 고객 분포')
plt.xlabel('지역')
plt.ylabel('고객 수')
plt.show()
```

### 성별 고객 분포

```python
# 성별 고객 분포
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data, palette='Set1')
plt.title('성별 고객 분포')
plt.xlabel('성별')
plt.ylabel('고객 수')
plt.show()
```

### 나이별 고객 분포

```python
# 나이 분포
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, color='skyblue')
plt.title('나이 분포')
plt.xlabel('나이')
plt.ylabel('빈도')
plt.show()
```

### 신용점수 분포

```python
# 신용 점수 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data['CreditScore'], bins=30)
plt.title('신용 점수 분포')
plt.xlabel('CreditScore')
plt.ylabel('Frequency')
plt.show()
```

### 거래 기간 분포

```python
# 거래 기간(Tenure) 분포 시각화
plt.figure(figsize=(10, 6))
sns.countplot(x='Tenure', data=data)
plt.title('거래 기간 분포')
plt.xlabel('Tenure (years)')
plt.ylabel('Count')
plt.show()
```

### 잔액 분포

```python
# 잔액(Balance) 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data['Balance'], bins=30, kde=True)
plt.title('잔액 분포')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.show()
```

### 잔액 0원은 제외한 잔액 분포

```python
# 0이 아닌 잔액 데이터만 필터링
non_zero_balance = data[data['Balance'] > 0]

# 필터링된 데이터로 잔액 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(non_zero_balance['Balance'], bins=30, kde=True)
plt.title('0이 아닌 잔액 분포')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.show()
```

### 고객 이탈 분포

```python
# 이탈 여부 시각화
plt.figure(figsize=(8, 5))
sns.countplot(x='Exited', data=data)
plt.title('은행 이탈 여부 분포')
plt.xlabel('Exited (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
!pip install matplotlib seaborn pandas

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# 예상 연봉(EstimatedSalary) 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data['EstimatedSalary'], bins=30, kde=True)
plt.title('예상 연봉 분포')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# 예상 연봉을 구간별로 나누기
salary_bins = pd.cut(data['EstimatedSalary'], bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000])
salary_bin_counts = salary_bins.value_counts().sort_index()

# 구간별 빈도수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=salary_bin_counts.index, y=salary_bin_counts.values)
plt.xticks(rotation=45)
plt.title('예상 연봉 구간별 분포')
plt.xlabel('Estimated Salary Range')
plt.ylabel('Frequency')
plt.show()
```
