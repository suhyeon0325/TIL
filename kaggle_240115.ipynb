```python
# 예상 연봉(EstimatedSalary) 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data['EstimatedSalary'], bins=30, kde=True)
plt.title('예상 연봉 분포')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# 예상 연봉을 구간별로 나누기
salary_bins = pd.cut(data['EstimatedSalary'], bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000])
salary_bin_counts = salary_bins.value_counts().sort_index()

# 구간별 빈도수 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=salary_bin_counts.index, y=salary_bin_counts.values)
plt.xticks(rotation=45)
plt.title('예상 연봉 구간별 분포')
plt.xlabel('Estimated Salary Range')
plt.ylabel('Frequency')
plt.show()
```

## 상관관계 분석

```python
# 상관관계 매트릭스 계산
correlation_matrix = data.corr()

# Exited와 다른 변수들 간의 상관관계 추출 (Exited 자체는 제외)
correlation_with_exited = correlation_matrix[['Exited']].drop('Exited').sort_values(by='Exited', ascending=False)

# 상관관계 히트맵 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_with_exited, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation with Exited')
plt.show()
```

```python
# 상관관계 매트릭스 계산
correlation_matrix = data.corr()

# Exited와 다른 변수들 간의 상관관계 추출 (Exited 자체는 제외)
correlation_with_exited = correlation_matrix['Exited'].drop('Exited').sort_values(ascending=False)

# 상관관계 막대 그래프 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_with_exited.index, y=correlation_with_exited.values)
plt.title('Correlation with Exited')
plt.xticks(rotation=45)
plt.ylabel('Correlation Coefficient')
plt.show()
```

## 거래기간과 활성회원

```python
# 'Tenure'와 'IsActiveMember' 간의 상관계수 계산
correlation_tenure_active = data[['Tenure', 'IsActiveMember']].corr()
print("Correlation between Tenure and IsActiveMember:\n", correlation_tenure_active)

# 거래 기간에 따른 활성 회원 비율 시각화
active_member_rate_by_tenure = data.groupby('Tenure')['IsActiveMember'].mean()

# 시각화
plt.figure(figsize=(10, 6))
active_member_rate_by_tenure.plot(kind='bar', color='skyblue')
plt.title('Active Member Rate by Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Active Member Rate')
plt.grid(True)
plt.show()
```

### 잔액 0원 분석

```python
# 잔액이 0원인 고객과 그렇지 않은 고객으로 데이터를 나눔
balance_zero = data[data['Balance'] == 0]
balance_nonzero = data[data['Balance'] > 0]

# 주요 변수들에 대한 기술 통계 비교
comparison_stats = pd.DataFrame({
    'Overall': data.describe().loc['mean'],
    'Balance = 0': balance_zero.describe().loc['mean'],
    'Balance > 0': balance_nonzero.describe().loc['mean']
})

print(comparison_stats)
```

```python
# 상관관계 매트릭스 계산
correlation_matrix = data.corr()

# Exited와 다른 변수들 간의 상관관계 추출 (Exited, id, CustomerId 자체는 제외)
correlation_with_exited = correlation_matrix['Exited'].drop(['Exited', 'id', 'CustomerId']).sort_values(ascending=False)

# 상관관계 막대 그래프 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_with_exited.index, y=correlation_with_exited.values)
plt.title('Correlation with Exited')
plt.xticks(rotation=45)
plt.ylabel('Correlation Coefficient')
plt.show()
```
