### Tutorial: How do I select a sybset of a DF?

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
ages = titanic["age"]
ages.head()
```

```python
type(titanic["age"])
```

```python
titanic["age"].shape
```

```python
age_sex = titanic[["age", "sex"]]
age_sex
```

```python
age_sex.head()
```

```python
type(titanic[["age", "sex"]])
```

```python
titanic[["age", "sex"]].shape
```

```python
# DF에서 원하는 행 추출하기
above_35 = titanic[titanic["age"] > 35]
above_35
```

```python
above_35.head()
```

```python
titanic["age"] > 35
```

```python
above_35.shape
```

```python
class_23 = titanic[titanic["pclass"].isin([2, 3])]
class_23
```

```python
class_23.head()
```

```python
# 위와 동일
class_23 = titanic[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)]
class_23
```

```python
# 나이 정보가 있는 승객만
# notna는 누락되지 않은 정보만 추출해줌
age_no_na = titanic[titanic["age"].notna()]
age_no_na.head()
```
